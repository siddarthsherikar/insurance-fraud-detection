```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv(r'C:\Users\M.ritika\OneDrive\Downloads\Copy of Worksheet in Case Study question 2.csv')
data.head()
```

```python
data.replace('?',np.nan,inplace=True)
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
missing=data.isnull().sum()/len(data)
missing=missing[missing>0]
missing.sort_values(inplace=True)
missing=missing.to_frame()
missing.columns=['Null_Count']
missing.index.names=['Col_Name']
missing=missing.reset_index()
sns.set(style='whitegrid',color_codes=True)
sns.barplot(x='Col_Name',y='Null_Count',data=missing)
plt.xticks(rotation=90)
plt.show()
```

```python
data['authorities_contacted']=data['authorities_contacted'].fillna(data['authorities_contacted'].mode()[0])
data['collision_type']=data['collision_type'].fillna(data['collision_type'].mode()[0])
data['police_report_available']=data['police_report_available'].fillna(data['police_report_available'].mode()[0])
```

```python
unique=data.nunique().to_frame()
unique.columns=['Count']
unique.index.names=['ColName']
unique=unique.reset_index()
sns.set(style='whitegrid',color_codes=True)
sns.barplot(x='ColName',y='Count',data=unique)
plt.xticks(rotation=90)
plt.show()
```

```python
#drop columns that are not used in project
to_drop=['policy_number','policy_bind_date','policy_state','insured_zip','incident_location','incident_date','incident_state','incident_city','insured_hobbies','auto_make','auto_model','auto_year']
data.drop(to_drop,inplace=True,axis=1)
```

```python
data.head()
```

```python
plt.figure(figsize=(18,15))
corr=data.select_dtypes(include='number').corr()
mask=np.triu(np.ones_like(corr,dtype=bool))
sns.heatmap(data=corr,mask=mask,annot=True,fmt='.2g',linewidth=1)
plt.show()
```

```python
data.drop(columns = ['age','total_claim_amount'],inplace=True,axis=1)
```

```python
data.head()
```

```python
#get target and independent features separated
x=data.drop('fraud_reported',axis=1)
y=data['fraud_reported']
```

```python
categorical_cols=x.select_dtypes(include=['object'])
categorical_cols=pd.get_dummies(categorical_cols,drop_first=True)
categorical_cols.head()
```

```python
numerical_col=x.select_dtypes(include=['int64'])
x=pd.concat([numerical_col,categorical_cols],axis=1)
```

```python
x.head()
```

```python
#outlier check
plt.figure(figsize=(20,15))
plotnumber=1
for col in x.columns:
    if plotnumber <=24:
        ax=plt.subplot(5,5,plotnumber)
        sns.boxplot(x[col])
        plt.xlabel(col,fontsize=15)
    plotnumber +=1
plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_trains,y_test=train_test_split(x,y,test_size=0.25)
```

```python
x_train.head()
```

```python
numerical_data=x_train[['months_as_customer','policy_deductable','umbrella_limit',
                        'capital-gains','capital-loss','incident_hour_of_the_day',
                        'number_of_vehicles_involved','bodily_injuries','witnesses','injury_claim','property_claim',
                        'vehicle_claim']]
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled_data=scaler.fit_transform(numerical_data)
```

```python
scaled_num_df=pd.DataFrame(data=scaled_data,columns=numerical_data.columns,index=x_train.index)
scaled_num_df.head()
```

```python
x_train.drop(columns=scaled_num_df.columns,inplace=True)
x_train=pd.concat([scaled_num_df,x_train],axis=1)
x_train.head()
```

```python
from sklearn.svm import SVC
svc_model=SVC()
svc_model.fit(x_train,y_trains)
y_pred=svc_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
svc_model_train_acc=accuracy_score(y_trains,svc_model.predict(x_train))
svc_model_test_acc=accuracy_score(y_test,y_pred)
print("Training Accurancy : " ,svc_model_train_acc)
print("Testing Accurancy : " ,svc_model_test_acc)
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_model=KNeighborsClassifier(n_neighbors=30)
knn_model.fit(x_train,y_trains)
y_pred=knn_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
knn_model_train_acc=accuracy_score(y_trains,svc_model.predict(x_train))
knn_model_test_acc=accuracy_score(y_test,y_pred)
print("Training Accurancy : " ,knn_model_train_acc)
print("Testing Accurancy : " ,knn_model_test_acc)
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
decision_tree_model=DecisionTreeClassifier()
decision_tree_model.fit(x_train,y_trains)
y_pred=decision_tree_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
decision_tree_model_train_acc=accuracy_score(y_trains,svc_model.predict(x_train))
decision_tree_model_test_acc=accuracy_score(y_test,y_pred)
print("Training Accurancy : " ,decision_tree_model_train_acc)
print("Testing Accurancy : " ,decision_tree_model_test_acc)
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
grid_params={
    'criterion':['gini','entropy'],
    'max_depth':[3,5,7,10],
    'min_samples_split':range(2,10,1),
    'min_samples_leaf':range(2,10,1)
}
grid_search=GridSearchCV(decision_tree_model,grid_params,cv=5,n_jobs=-1,verbose=1)
grid_search.fit(x_train,y_trains)
```

```python
print(grid_search.best_params_)
print(grid_search.best_score_)
```

```python
decision_tree_model=grid_search.best_estimator_
y_pred=decision_tree_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
decision_tree_model_train_acc=accuracy_score(y_trains,svc_model.predict(x_train))
decision_tree_model_test_acc=accuracy_score(y_test,y_pred)
print("Training Accurancy : " ,decision_tree_model_train_acc)
print("Testing Accurancy : " ,decision_tree_model_test_acc)
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model=RandomForestClassifier(criterion='entropy',max_depth=15,max_features='sqrt',min_samples_leaf=1,min_samples_split=3,n_estimators=14)
rf_model.fit(x_train,y_trains)
y_pred=rf_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
rf_model_train_acc=accuracy_score(y_trains,svc_model.predict(x_train))
rf_model_test_acc=accuracy_score(y_test,y_pred)
print("Training Accurancy : " ,rf_model_train_acc)
print("Testing Accurancy : " ,rf_model_test_acc)
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
models=pd.DataFrame({
    'Model' :['svc_model','knn_model','Decision Tree','Random Forest'],
    'Score' :[svc_model_test_acc, knn_model_test_acc,decision_tree_model_test_acc,rf_model_test_acc]
})
models.sort_values(by='Score',ascending=False)
```

```python
plt.bar(models['Model'],models['Score'],width=0.4,color='maroon')
plt.xlabel("Models")
plt.xticks(rotation=90)
plt.ylabel("Scores")
plt.title("Models for Insurance Fraud Detection")
plt.show()
```
